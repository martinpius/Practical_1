Data Visualization in Python
 ***This can be done using Matplotlib or Seabon***

*Data Visualization using Matplotlib*

**Figure**: This is the whole region of space that’s created when you create any figure. The Figure object is the overall object that contains everything else.
**Axes**: An Axes object is the object that contains the x-axis and y-axis for a 2D plot. Each Axes object corresponds to a plot or a graph. You can have more than one Axes object in a Figure, as you’ll see later on in this Chapter.
**Axis**: An Axis object contains one of the axes, the x-axis or the y-axis for a 2D plot.

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
plt.show()
```

```python
import matplotlib.pyplot as plt
days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
steps_walked = [8934, 14902, 3409, 25672, 12300, 2023, 6890]
steps_last_week = [9788, 8710, 5308, 17630, 21309, 4002, 5223]
fig, ax = plt.subplots()
ax.plot(days, steps_walked, "o-g")
ax.plot(days, steps_last_week, "x--m")
ax.set_title("Step count | This week and last week")
ax.set_xlabel("Days of the week")
ax.set_ylabel("Steps walked")
ax.grid(True)
ax.legend(["This week", "Last week"], loc = "upper left")
plt.show()
plt.savefig('figure1.png', format='png', dpi=300)### eps, svg etc
```

**Creating Subplots**

```python
import matplotlib.pyplot as plt
fig, axs = plt.subplots(nrows = 1, ncols = 2)
plt.show()
```

**Therefore, plt.subplots() returns one of the following return values:

a tuple containing a Figure object and an Axes object if nrows and ncols are both 1. In this case, only one Axes object is created
a tuple containing a Figure object and a numpy.ndarray if nrows and ncols aren’t both 1. The array contains Axes objects. The shape of the array is the same as the shape of the grid containing the subplots. Therefore, the call plt.subplots(2, 2), for example, will return a 2x2 numpy.ndarray as its second return value.**

```python
import matplotlib.pyplot as plt
days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
steps_walked = [8934, 14902, 3409, 25672, 12300, 2023, 6890]
steps_last_week = [9788, 8710, 5308, 17630, 21309, 4002, 5223]
fig, axs = plt.subplots(1, 2)
axs[0].plot(days, steps_walked, "o-g")
axs[0].plot(days, steps_last_week, "v--m")
axs[0].set_title("Step count | This week and last week")
axs[0].set_xlabel("Days of the week")
axs[0].set_ylabel("Steps walked")
axs[0].grid(True)
plt.show()
plt.savefig('figure3.png', format='png', dpi=300)### eps, svg etc
```

***Seaborn Plot types
Seaborn provides a wide range of plot types that can be used for data visualization and exploratory data analysis. Broadly speaking, any visualization can fall into one of the three categories.

Univariate – x only (contains only one axis of information)

Bivariate – x and y (contains two axis of information)

Trivariate – x, y, z (contains three axis of information)***

**ADVANCED VISUALIZATION IN PYTHON**

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load built-in dataset
df = sns.load_dataset("tips")
```

**FACEGRID**

```python
g = sns.FacetGrid(df, col="sex", row="time", margin_titles=True)
g.map(sns.histplot, "total_bill", bins=10, kde=True)
plt.show()
```

**HEARMAP----Correlation Matrix**

**PAIRED PLOTS**

```python
sns.pairplot(df, hue="sex", diag_kind="kde")
plt.suptitle("Pairwise Relationships", y=1.02)
plt.show()
```

**Violin Plot + Swarm Plot)**

```python
sns.violinplot(x="day", y="total_bill", data=df, inner=None)
sns.swarmplot(x="day", y="total_bill", data=df, color="k", alpha=0.6)
plt.title("Violin + Swarm: Total Bill by Day")
plt.show()
```

**Catplot – Advanced Categorical Plotting**

```python
sns.catplot(x="day", y="tip", hue="smoker", col="sex", kind="box", data=df)
plt.suptitle("Tips by Day, Gender and Smoking Status", y=1.02)
plt.show()
```

**Regression Plot – Trend with Confidence Interval**

```python
sns.lmplot(data=df, x="total_bill", y="tip", hue="smoker", col="sex", aspect=1)
plt.suptitle("Linear Fit by Smoking Status and Gender", y=1.02)
plt.show()
```

**Custom Color Palette**

```python
palette = sns.color_palette("Set2")
sns.boxplot(x="day", y="tip", hue="sex", data=df, palette=palette)
plt.title("Custom Colors: Tip by Day and Sex")
plt.show()
```
